```python
import pandas as pd
import os
import numpy as np
```

```python
data_training_awal= pd.read_excel('D:\\JAMALI FORECASTING\\B_train_data.xlsx', index_col=0)
data_training= data_training_awal[47:].astype('float32').copy()
data_training
```

```python
last_data_training= data_training.tail(1).copy()
data_training= data_training[:-1].copy()
last_data_training
```

```python
data_training
```

```python
df = data_training.copy()

# 1. Hitung sudut kembali menggunakan arctan2(sin, cos)
# Fungsi ini mengembalikan nilai sudut dalam satuan radian
angles = np.arctan2(df['Hari_sin'], df['Hari_cos'])

# 2. Kembalikan ke angka hari dalam rentang 0 hingga 6
# Operasi modulo 7 memastikan angka negatif dari arctan2 kembali ke putaran yang benar
df['Hari_Indeks'] = np.round((angles * 7) / (2 * np.pi)).astype(int) % 7
```

```python
df['Hari_Indeks'].unique()
```

```python
df_senin= df[df['Hari_Indeks'] ==0]
df_senin = df_senin.drop(columns=['Hari_Indeks']).copy()
df_senin
```

```python
# Daftar nama hari untuk penamaan file
nama_hari = ['senin', 'selasa', 'rabu', 'kamis', 'jumat', 'sabtu', 'minggu']

# Loop untuk memproses setiap hari berdasarkan indeks 0 sampai 6
for i, hari in enumerate(nama_hari):
    # 1. Filter data berdasarkan indeks hari (0=Senin, 1=Selasa, dst)
    df_temp = df[df['Hari_Indeks'] == i].copy()
    
    # 2. Hapus kolom 'Hari_Indeks' agar file excel lebih bersih seperti contoh Anda
    df_temp = df_temp.drop(columns=['Hari_Indeks'])
    
    # 3. Simpan ke Excel dengan format nama yang diminta
    nama_file = f'D:\\JAMALI FORECASTING\\PER_HARI\\DATA_HARI\\data_{hari}_train.xlsx'
    df_temp.to_excel(nama_file, index=True)
    
    print(f'Berhasil menyimpan: {nama_file}')
```
