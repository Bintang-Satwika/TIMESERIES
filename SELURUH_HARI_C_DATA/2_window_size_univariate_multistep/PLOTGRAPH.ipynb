```python
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker 

# 1. Define the specific files you want to load
file_names = [
    "metrics_eval48.json",
    "metrics_eval480.json",
    "metrics_eval672.json"
]

# 2. Initialize a dictionary to hold the merged data
# We assume the structure contains "MSE_train" and "MSE_val"
merged_metrics = {
    "MSE_train": [],
    "MSE_val": []
}

# 3. Import and append data
for file in file_names:
    try:
        with open(file, 'r') as f:
            data = json.load(f)
            # Appending the values to our lists
            merged_metrics["MSE_train"].append(data["MSE_train"])
            merged_metrics["MSE_val"].append(data["MSE_val"])
    except FileNotFoundError:
        print(f"Warning: {file} not found. Skipping.")

# Now 'merged_metrics' is ready to be used in your plotting code

# 4. Optional: Save the merged file
# with open('merged_metrics.json', 'w') as f:
#     json.dump(merged_metrics, f)
# metrics_dict_1= 
# # Extract data
data_train = merged_metrics["MSE_train"]
data_train =[item for sublist in data_train for item in sublist]
data_val = merged_metrics["MSE_val"]
data_val =[item for sublist in data_val for item in sublist]

print(len(data_train), len(data_val))
# # Generate x-values dynamically
start_step=24
end_step=24*28
step_slice=24
x_values = np.arange(start_step, end_step+1, step_slice)

# Plot
plt.figure(figsize=(12, 5))
plt.plot(x_values, data_train, marker='o', linestyle='-', color='b', label="MSE Train")
plt.plot(x_values, data_val, marker='s', linestyle='--', color='r', label="MSE Validation")
plt.axvline(x=(np.argmin(data_val)+1)*step_slice, color='r', linestyle='--', linewidth=2, label="Lowest MSE Validation")
# Labels
plt.xlabel("window size (timestep)")
plt.ylabel("MSE Value")
plt.title("MSE Train vs Validation Over Steps")
plt.xticks(np.arange(12,end_step+999, 24),fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(0, end_step+1)
plt.legend()

# Show plot
plt.show()

print("MSE best train")
print((np.argmin(data_train)+1)*24, np.min(data_train))
print("MSE best val")
print((np.argmin(data_val)+1)*24, np.min(data_val))
```
