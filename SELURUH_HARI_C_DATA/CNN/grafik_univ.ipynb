```python
import json
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os

current_dir = os.getcwd()

# Load data
file_path = os.path.join(current_dir, 'loss_history_CNN_univ_1_20_epoch50.json')
with open(file_path, 'r') as file:
    history = json.load(file)

start_epoch = 1
end_epoch = 25
training_loss = history['loss'][start_epoch:end_epoch]
validation_loss = history['val_loss'][start_epoch:end_epoch]
epochs = range(start_epoch, end_epoch)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# Plot pertama: Training Loss
ax1.plot(epochs, training_loss, 'b', label='Training loss')
ax1.set_title('Training Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss Value')

# 1. Mengatur rentang sumbu y (Batas Bawah, Batas Atas)
ax1.set_ylim(0.004, 0.024) 

# 2. Mengatur kelipatan sumbu y (Misal: setiap 0.002)
ax1.yaxis.set_major_locator(ticker.MultipleLocator(0.002))

# Mengatur sumbu x tetap setiap 2 unit
ax1.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax1.set_xlim(left=start_epoch)
ax1.legend()

# Plot kedua: Validation Loss
ax2.plot(epochs, validation_loss, 'b', label='Validation loss')
ax2.set_title('Validation Loss')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss Value')

# Mengatur rentang dan kelipatan sumbu y untuk plot kedua
ax2.set_ylim(0.1, 1.0)
ax2.yaxis.set_major_locator(ticker.MultipleLocator(0.05))

ax2.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax2.set_xlim(left=start_epoch)
ax2.legend()

plt.tight_layout()
plt.show()
```
