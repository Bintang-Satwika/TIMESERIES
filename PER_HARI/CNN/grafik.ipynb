```python
import json
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker # Tambahkan import ini
import os

current_dir = os.getcwd()
print(f"Direktori saat ini: {current_dir}")

# Load data
file_path = os.path.join(current_dir, 'loss_history_CNN_monday_20.h5_end.json')
with open(file_path, 'r') as file:
    history = json.load(file)

start_epoch = 1
training_loss = history['loss'][start_epoch:]
validation_loss = history['val_loss'][start_epoch:]

# Range sumbu x
epochs = range(start_epoch, start_epoch + len(training_loss))

# Membuat figure dengan 2 subplot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10))

# Plot pertama: Training Loss
ax1.plot(epochs, training_loss, 'b', label='Training loss')
ax1.set_title('Training Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss Value')
ax1.set_xlim(left=start_epoch)

# Mengatur interval sumbu x setiap 2 unit
ax1.xaxis.set_major_locator(ticker.MultipleLocator(2)) 
ax1.legend()

# Plot kedua: Validation Loss
ax2.plot(epochs, validation_loss, 'b', label='Validation loss')
ax2.set_title('Validation Loss')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss Value')
ax2.set_xlim(left=start_epoch)
ax2.set_ylim(0.03, 0.2)
ax2.yaxis.set_major_locator(ticker.MultipleLocator(0.01))
# Mengatur interval sumbu x setiap 2 unit
ax2.xaxis.set_major_locator(ticker.MultipleLocator(2)) 
ax2.legend()

plt.tight_layout()
plt.show()
```
